## Agrupando dados

### Função groupby()

````python
df.head()
````

````python
df.groupby('nome_loja')['valor_final'].sum().to_frame()
````

````python
df_faturamento_funcionario = df.groupby('nome_funcionario')['preco_custo'].sum().to_frame()
````

````python
df_faturamento_funcionario.loc['Pamela Lima']
````

### Função agg()

````python
df.groupby("descricao_produto")["valor_final"].agg(['sum', 'mean', 'min', 'max'])
````

````python
df_agregado = df.groupby('descricao_produto').agg({
    'preco_custo': 'sum',
    'valor_final': 'mean'
})
````

### Agrupando dados pelo índice

````python
df.groupby(df.index).agg({'valor_final': 'sum'})
````

### Agrupando dados por mais de uma coluna

````python
df.groupby(['nome_loja', 'nome_funcionario', 'descricao_produto']).agg({'valor_final': 'sum'})
````

### Agrupando dados por tabelas dinâmicas (Pivot Tables)

```python
pivot_table()
```

````python
df.pivot_table(
    index='nome_funcionario',
    values=['valor_final'],
    aggfunc='sum',
    columns='descricao_produto',
    margins=True
)
````

## Iterando sobre um DataFrame

### iterrows()

- retorna o index (label) e o conteúdo da linha (row) como uma Series

````python
for index, row in df.iterrows():
    print(index)       
    print(row)
    break
    
````

### itertuples()

- retorna uma tupla nomeada


````python
for row in df.itertuples():
    print(f'Categoria do produto: {row.categoria}')
    print(f'Descrição do produto: {row.descricao_produto}')
    print(f'Valor final: {row.valor_final}')
    print('\n')
    
````
